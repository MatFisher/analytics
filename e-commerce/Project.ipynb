Создадим функцию для определения причины:

```python
#define function for classifying orders based on statuses

def define_reason(row):
 if row['order_status'] == 'shipped':
     val = 'Товар отгружен со склада'
 elif row['order_status'] == 'canceled':
     val = 'Заказ отменен'
 elif row['order_status'] == 'unavailable':
     val = 'Заказ недоступен'
 elif row['order_status'] == 'invoiced':
     val = 'Заказ еще не оплачен, выставлен счет'
 elif row['order_status'] == 'processing':
     val = 'Заказ еще в процессе сборки'
 elif row['order_status'] == 'delivered':
     val = 'Возможна ошибка, ПРОВЕРИТЬ ЭТИ ЗАКАЗЫ'
 elif row['order_status'] == 'created':
     val = 'Заказ в статусе создан'
 elif row['order_status'] == 'approved':
     val = 'Заказ подтвержден, еще не оплачен'
 return val
```

Теперь создаем столбец, где будет указана причина того, что заказ пока не доставлен

```python
#create new column 'non_delivered_reason' using the function above
non_delivered_orders['non_delivered_reason'] = non_delivered_orders.apply (define_reason, axis=1)
```

Посмотрим на результат, столбец добавился, можно анализировать дальше.

```python
non_delivered_orders['order_purchase_month'] = non_delivered_orders['order_purchase_timestamp'].dt.to_period("M")
```

Для дальнейших рассчетов нам понадобится знать год и номер месяца, для расчета количества дней

```python
purch2items['order_purchase_year'] = purch2items['order_purchase_timestamp'].dt.year
purch2items['order_purchase_month'] = purch2items['order_purchase_timestamp'].dt.month
purch2items['order_purchase_day'] = purch2items['order_purchase_timestamp'].dt.day
```

```python
purch2items.head()
```

Проверю, целиком ли учтены месяцы в наших датасетах, нет ли такого, что в каком-то месяце не хватает дней.

```python
purch2items.groupby(by=['order_purchase_year','order_purchase_month'], as_index = False) \
           .agg({'order_purchase_day':'nunique'}) \
           .sort_values(by=["order_purchase_year","order_purchase_month"])
```

Итак, видно, что в основном месяцы полные.

Однако, сентябрь, октябрь, декабрь 2016 года заполнено лишь частично, данных за ноябрь 2016 и вовсе нет.
Январь 2017 - не полный месяц, всего 27 дней из 31.

Август 2018 года - не полный, всего 29 дней из 31.

За сентябрь 2018 запись есть только за 1 день.

Учтем это при определении количества дней в месяце.

```python
def days_in_month(row):
 if row['order_purchase_year'] == '2016' and row['order_purchase_month'] == 9:
     val = 2
 elif row['order_purchase_year'] == '2016' and row['order_purchase_month'] == 10:
     val = 8
 elif row['order_purchase_year'] == '2016' and row['order_purchase_month'] == 12:
     val = 1
 elif row['order_purchase_year'] == '2017' and row['order_purchase_month'] == 1:
     val = 27
 elif row['order_purchase_year'] == '2018' and row['order_purchase_month'] == 8:
     val = 29
 elif row['order_purchase_year'] == '2018' and row['order_purchase_month'] == 9:
     val = 1
 elif row['order_purchase_month'] in (1,3,5,7,8,10,12):
     val = 31
 elif row['order_purchase_month'] in (4,6,9,11):
     val = 30
 elif row['order_purchase_month'] == 2:
     val = 28
 else:
     val = 0 
 return val
```

```python
purch2items['days_in_month'] = purch2items.apply(lambda x: days_in_month(x), axis=1)
```

Расчет количества недель в месяце:

```python
purch2items['weeks_in_month'] = purch2items["days_in_month"] / 7
```

```python
purch2items = purch2items.groupby(['order_purchase_month','customer_unique_id','weeks_in_month']) \
                         .agg({'order_id':'nunique'}).reset_index() \
                         .rename(columns={"order_id": "orders_count"})
```

```python
purch2items['mean_cnt_orders'] = purch2items.orders_count / purch2items.weeks_in_month
```

## 6. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию.

Часто для качественного анализа аудитории используют подходы, основанные на сегментации.  

В кластеризации можешь выбрать следующие метрики: 

R - время от последней покупки пользователя до текущей даты, 

F - суммарное количество покупок у пользователя за всё время, 

M - сумма покупок за всё время. 

Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. 

Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю. 

Описание подхода можно найти тут.

Для данного пункта нам понадобится датасет с покупками purchases + нужны данные по составу каждого заказа - olist_order_items_dataset

```python
purch2items_6 = purchases \
        .merge(order_items, how = 'inner', on = 'order_id') \
        .merge(customers, how = 'inner', on = 'customer_id') \
        [['order_id', 'order_approved_at', 'customer_unique_id', 'order_purchase_timestamp', 'order_item_id', 'product_id', 'price']]
```

Сейчас 2024 год, и если мы будем брать текущую дату как точку отсчета, чтобы рассчитать показатель **recency** - выйдет некорректно. Поэтому от последней даты покупки отсчитаем 1 день, и будем от этой даты отсчитывать давность покупок. <br>

Буду использовать дату *order_purchase_timestamp*.

```python
snapshot_date = purch2items_6['order_purchase_timestamp'].max() + timedelta(days=1)
print(snapshot_date)
```

Теперь можно рассчитать показатели. <br>
<br>
**Recency** - это количество дней с последней покупки до даты *snapshot_date*<br>
**Frequency** - это количество покупок, в нашем случае - это количество заказов (нужно считать уникальные *order_id*).<br>
**Monetary** - это сумма по столбцу *price*.<br>
   

```python
rfm_processing = purch2items_6.groupby(['customer_unique_id']).agg({ \
             'order_purchase_timestamp': lambda x: (snapshot_date - x.max()).days, \
             'order_id': 'nunique',  \
             'price': 'sum' })

rfm_processing.rename(columns={'order_purchase_timestamp':'recency', \
                               'order_id':'frequency', \
                               'price':'monetary'}, inplace=True)
```

```python
print(rfm_processing.head())
print('В датафрейме {} строк и {} столбцов.' \
     .format(rfm_processing.shape[0], rfm_processing.shape[1]))
```

Итак, у нас есть 94970 пользователей, которых нужно сегментировать с помощью методов RFM-анализа. <br>

Нужно разбить показатели **recency**, **frequency** и **monetary** на бакеты. 
По сути, каждому значению мы присвоим балл, который показывает нам, насколько "хорошее" это для нас значение.
Будем оценивать показатели от 1 до 5. <br>
<br>
**recency** - как давно что-то покупал? Чем более недавно, тем лучше<br>
**frequency** - как часто покупал? Чем чаще, тем лучше<br>
**monetary** - как много денег у нас потратил? Чем больше, тем лучше<br>

Воспользуемся методом **qcut** <br>

```python
pd.qcut(rfm_processing['recency'],5)
```

Показатель **recency** будем оценивать так: <br>
+ (0.999, 98.0] 5 баллов, так как это самые недавние по времени заказы <br>
+ (98.0, 182.0] 4 балла <br>
+ (182.0, 273.0] 3 балла <br>
+ (273.0, 388.0] 2 балла <br>
+ (388.0, 729.0] 1 балл, это самые давние покупки <br>

Проставляем баллы <br>

```python
rfm_processing['r_score'] = 6 - (pd.qcut(rfm_processing['recency'],5,labels=False) + 1)
```

Убедимся, что наша задача выполнена:

```python
rfm_processing.groupby('r_score')['recency'].agg(['mean', 'count'])
```

Действительно разделили показатели на 5 приблизительно равных групп. <br>
<br>

И учли тот факт, что в случае **recency** чем больше показатель, тем ниже балл, потому что большое значение **recency** означает, что покупатель был у нас очень давно.

С показателем **frequency** этот номер не пройдет. <br>
Слишком часто повторяются одинаковые значения, и мы не сможем разбить их на группы с помощью *qcut*. <br>
Продемонстрируем это:

```python
rfm_processing['frequency'].value_counts()
```

## И здесь придется применять индивидуальный подход для формирования бакетов.<br>
Разобъем эти значения на группы вручную.<br>
<br>
1 балл я поставлю тем пользователям, которые совершили только одну покупку. <br>
2 балла я поставлю тем пользователям, у которых 2 покупки.<br>
3 балла я поставлю пользователям, которые сделали 3-4 покупки.<br>
4 балла получат пользователи, у которых от 5-6 покупок.<br>
5 баллов получат пользователи, у которых больше или равно 7 покупок.<br>


```python
def get_f_score(x):
    if x == 1:
        return 1
    if x == 2:
        return 2
    if (x >= 3) & (x <= 4):
        return 3
    if (x >= 5) & (x <= 6):
        return 4
    if x >= 7:
        return 5
```

```python
rfm_processing['f_score'] = rfm_processing['frequency'].apply(get_f_score)
```

```python
rfm_processing.groupby('f_score')['frequency'].agg(['mean','count'])
```

Теперь приступим к показателю **monetary**. <br>
Также будем использовать метод **qcut**.

```python
pd.qcut(rfm_processing['monetary'],5)
```

Оценивать **monetary** будем так:
+ (0.849, 39.9] - 1 балл
+ (39.9, 69.9] - 2 балла
+ (69.9, 109.9] - 3 балла
+ (109.9, 179.9] - 4 балла
+ (179.9, 13440.0] - 5 баллов

Проставляем баллы:

```python
rfm_processing['m_score'] = pd.qcut(rfm_processing['monetary'],5,labels=False) + 1
```

Проверим:

```python
rfm_processing.groupby('m_score')['monetary'].agg(['mean','count'])
```

Можно рассчитать итоговую оценку:

```python
rfm_processing['rfm_score'] = rfm_processing['r_score'].map(str) \
                            + rfm_processing['f_score'].map(str) \
                            + rfm_processing['m_score'].map(str)
```

```python
rfm_processing.sort_values('rfm_score', ascending=False)
```

Получили список пользователей, упорядоченный по *rfm_score*, самые "лучшие" покупатели - в начале списка. <br>
А в конце списка - "оттекающие" покупатели. Они делали покупки очень давно, покупок было очень мало, и потратили они немного.

Теперь разобъем пользователей на сегменты. <br>

+ **Новички** - в эту категорию включим покупателей, которые сделали 1 покупку в течение последних 98 дней
+ **Дрейфующие** - сюда попадут покупатели, которые делали 1 заказ в период от 98 до 273 дней или от 2 до 4 заказов от 182 до 273 дней назад
+ **Спящие** - от 1 до 4 заказов, ранее чем 273 дня назад
+ **Перспективные** - от 2 до 4 заказов, 182 и меньше дней назад
+ **Лояльные** - от 5 и больше покупок, ранее чем 182 дня назад
+ **В зоне риска** - от 5 и больше покупок, от 182 до 388 дней назад
+ **В зоне потери** - от 5 и больше покупок, больше чем 273 дня назад

```python
segt_map = {
    r'51': 'новички',
    r'[3-4]1': 'дрейфующие',
    r'3[2-3]': 'дрейфующие',
    r'[1-2][1-3]': 'спящие',
    r'[4-5][2-3]': 'перспективные',
    r'[4-5][4-5]': 'лояльные',
    r'[2-3][4-5]': 'в зоне риска',
    r'[1][4-5]': 'в зоне потери'}    

rfm_processing['Segment'] = rfm_processing['r_score'].map(str) + rfm_processing['f_score'].map(str)
rfm_processing['Segment'] = rfm_processing['Segment'].replace(segt_map, regex=True)
rfm_processing.head()
```

```python
rfm_level_agg = rfm_processing.groupby('Segment').agg({
    'recency' : ['mean'],
    'frequency' : ['mean'],
    'monetary' : ['mean','count']}).round(1)
```

Получаем вот такое распределение пользователей по сегментам:

```python
rfm_level_agg.columns = rfm_level_agg.columns.droplevel()
rfm_level_agg.columns = ['Ср_дней_с_посл_покупки','Ср_кол-во_покупок','Ср_выручка','Кол-во_чел']
rfm_level_agg.sort_values(by='Кол-во_чел', ascending=False)
```
